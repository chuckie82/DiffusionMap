```python
import numpy as np
import pyqtgraph as qg
from numba import jit, float64, int32
import time
import matplotlib.pyplot as plt
```

```python
x = np.random.rand(int(10e6))
y = np.random.rand(int(10e6))
```

```python
x_edge = np.linspace(start=0, stop=1, num=800, endpoint=True)
y_edge = np.linspace(start=0, stop=1, num=800, endpoint=True)
```

```python
tic = time.time()
heatmap, xedges, yedges = np.histogram2d(x, y, bins=(x_edge,y_edge))
toc = time.time()
print(toc-tic)
```

```python
plt.figure(figsize=(10,10))
plt.imshow(heatmap)
plt.colorbar()
plt.show()
```
